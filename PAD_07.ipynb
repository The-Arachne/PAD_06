# ZADANIE 1


```python
import pandas as pd

data = pd.read_csv("wyniki.csv")
data.head
```

```python
Fdata = data[data['plec'] == 'F']
Mdata = data[data['plec'] == 'M']

Mdata.head
```

```python
import numpy as np

Mresults = Mdata['ocena_koncowa']
M_STD = np.std(Mresults)
M_MEAN = np.mean(Mresults)
M_SHAPE = Mresults.shape[0]

Fresults = Fdata['ocena_koncowa']
F_STD = np.std(Fresults)
F_MEAN = np.mean(Fresults)
F_SHAPE = Fresults.shape[0]


T_STAT = (M_MEAN - F_MEAN)/(np.sqrt(np.square(M_STD)/M_SHAPE + np.square(F_STD)/F_SHAPE))
T_STAT
```

```python
import scipy.stats as stats

df = data.shape[0]-1

p = stats.t.sf(abs(T_STAT), df=df)
p
```

```python
# p value jest mniejsze 0.05 -> hipoteza zerowa do odrzucenia
# hipoteza alternatywna jest prawid≈Çowa
```

# ZADANIE 2

```python
data = pd.read_csv('ZyskiFirmyX.csv')
data.head
```

```python
data['Rok'].describe()
```

```python
import plotly.express as px

px.scatter(data, "Rok", "Zysk")
```

```python
import statsmodels.formula.api as stsfmapi

model = stsfmapi.ols(formula="Zysk ~ Rok", data=data).fit()
print(model.summary())
```

```python
import plotly.graph_objects as go

data["Fit"] = model.fittedvalues

fig = go.Figure()

fig.add_trace(
    go.Scatter(x=data["Rok"], y=data["Zysk"], name="Rok vs Zysk")
)
fig.add_trace(
    go.Scatter(x=data["Rok"], y=data["Fit"], name="Trend - Lin. Reg.")
)
fig.update_layout(title="Regresja liniowa Zysku na przestrzeni lat", xaxis_title="Rok", yaxis_title="Zysk")
fig.show()
```

```python
data = pd.read_csv('spendingscores.csv')
data.head
```

```python
data['Gender'].replace(['Male', 'Female'], [1, 0], inplace=True)

data.head
```

```python
model = stsfmapi.ols("SpendingScore ~ Gender + Age + AnnualIncome", data=data).fit()
print(model.summary())
```

```python
fig = px.imshow(data.corr())
fig.show()
```

```python
#BEZ -> GENDER
model = stsfmapi.ols("SpendingScore ~ Age + AnnualIncome", data=data).fit()
print(model.summary())
```

```python
#BEZ -> AGE
model = stsfmapi.ols("SpendingScore ~ Gender + AnnualIncome", data=data).fit()
print(model.summary())
```

```python
#BEZ -> INCOME
model = stsfmapi.ols("SpendingScore ~ Gender + Age", data=data).fit()
print(model.summary())
```
